```python
median_ages = np.zeros((2,3))
median_ages
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df[(df['Gender'] == i) & \
                              (df['Pclass'] == j+1)]['Age'].dropna().median()
 
median_ages
```

```python
df['AgeFill'] = df['Age']

df.head()
```
